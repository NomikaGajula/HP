```python
y_train=data.Result
x_train=data.iloc[:,:-1]
x_test=data1.iloc[:,:]
X_test=data2.iloc[:,:]
```

```python
x_test
```

```python
y_train
```

Models

1. DecisionTree
2. Random Forest

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
model=DecisionTreeClassifier()
```

```python
#fitting the training values in to the model
model.fit(x_train,y_train)
```

```python
#predicting the y value by taking the x_test values
y_predict=model.predict(x_test)
```

```python
y_predict
```

```python
#Adding the result column to the data
Data1=pd.DataFrame(data1.key)
Data1.loc[:,'result']=y_predict
```

```python
Data1.head()
```

```python
#counting no of phishing values and non-phishing values(-1 and 1 values)
pd.value_counts(Data1.result)
```

```python
#converting the data file into the csv file
Data1.to_csv(r'D:\NOMI FOLDER\export_dataframe3(Decision tree).csv')
```
